### 영문 컬럼명 -> 한글 컬럼명으로 변환하기

```python
### 직접 컬럼명을 딕셔너리로 만들어서 사용할 경우
"""
{'on_date': '승차시각',
 'off_date': '하차시각',
 'route_name': '노선명',
 'descr': '노선설명',
 'age_type': '승객연령',
 'trans_yn': '환승여부',
 'addfee_yn': '추가운임여부',
 'start_bstop': '승차정류장',
 'start_gps_x': '승차정류장 GPS X',
 'start_gps_y': '승차정류장 GPS Y',
 'end_bstop': '하차정류장',
 'end_gps_x': '하차정류장 GPS X',
 'end_gps_y': '하차정류장 GPS Y'}
"""
```

```python
### 데이터 정의서에서 "영문 및 한글" 컬럼명만 읽어들이기
# 컬럼명의 위치 : 엑셀기준 3번째 행
# 필요한 데이터의 열 : B, C
# 데이터프레임 변수명 : df_bus_card_col_org
file_path = "./01_data_원본/trfcard (0)/trfcard_columns.xlsx"
df_bus_card_col_org = pd.read_excel(file_path,
                                    header = 2,
                                    usecols = "B:C")
df_bus_card_col_org
```

```python
### 조회한 영문 및 한글컬럼 데이터를 이용해서
# 딕션너리 데이티로 변환하기
#  - key는 영문명, value는 한글명으로
#  - 변수명 : df_bus_card_col_new_dict

### 딕셔너리 만들기 - 1방식
df_bus_card_col_new_dict = {}
for k, v in zip(df_bus_card_col_org["컬럼명 (영문)"],
                df_bus_card_col_org["컬럼명 (한글)"]) :
    df_bus_card_col_new_dict[k] = v

### 딕셔너리 만들기 - 2방식
df_bus_card_col_new_dict = \
            dict(zip(df_bus_card_col_org["컬럼명 (영문)"],
                     df_bus_card_col_org["컬럼명 (한글)"]))

df_bus_card_col_new_dict
```

```python
### 원본 데이터프레임의 컬럼명을 한글로 변환하기
df_bus_card_kor = \
        df_bus_card_org.rename(columns = df_bus_card_col_new_dict)
```

```python
### 승차시각 및 하차시각 타입변환하가 : 문자형태로...
df_bus_card_kor = df_bus_card_kor.astype({"승차시각" : "str"})
df_bus_card_kor = df_bus_card_kor.astype({"하차시각" : "str"})
df_bus_card_kor.info()
```

```python
### 분석에 필요한 요소(컬럼)을 별도 데이터프레임으로 추출하기
df_bus_cart = pd.DataFrame()

df_bus_cart["승차시각"] = \
    pd.to_datetime(df_bus_card_kor["승차시각"])
df_bus_cart["하차시각"] = \
    pd.to_datetime(df_bus_card_kor["하차시각"])
df_bus_cart["승객구분"] = df_bus_card_kor["승객연령"]
df_bus_cart["환승여부"] = df_bus_card_kor["환승여부"]
df_bus_cart["추가운임여부"] = df_bus_card_kor["추가운임여부"]
df_bus_cart["승차정류장"] = df_bus_card_kor["승차정류장"]
df_bus_cart["하차정류장"] = df_bus_card_kor["하차정류장"]

df_bus_cart
```

```python
### 체류시간 컬럼 생성하기
# - 체류시간 = 하차시각 - 승차시각
df_bus_cart_temp = df_bus_cart.copy()
df_bus_cart_temp
```

```python
df_bus_cart_temp["체류시간"] = \
    df_bus_cart_temp["하차시각"] - df_bus_cart_temp["승차시각"]
df_bus_cart_temp.info()
```

```python
df_bus_cart_temp["체류시간"] = \
    (df_bus_cart_temp["하차시각"] - \
         df_bus_cart_temp["승차시각"]).astype("timedelta64[m]").astype(int)
df_bus_cart_temp
```

```python
df_bus_cart["버스내체류시간(분)"] = \
    (df_bus_cart["하차시각"] - \
         df_bus_cart["승차시각"]).astype("timedelta64[m]").astype(int)
df_bus_cart.info()
df_bus_cart.describe()
```

```python
### 기준년도, 기준월, 기준일, 기준시간, 기준시간(분) 컬럼생성
df_bus_cart["기준년도"] = df_bus_cart["승차시각"].dt.year
df_bus_cart["기준월"] = df_bus_cart["승차시각"].dt.month
df_bus_cart["기준일"] = df_bus_cart["승차시각"].dt.day
df_bus_cart["기준시간"] = df_bus_cart["승차시각"].dt.hour
df_bus_cart["기준시간(분)"] = df_bus_cart["승차시각"].dt.minute
df_bus_cart
```

### 모든 데이터 통합하기

```python
from datetime import datetime
```

```python
### 실행 시간 체크하기
# - 시작 시작
start_date = datetime.today().strftime("%Y-%m-%d %H:%M:%S")

### 최종 통합 데이터를 저장할 데이터프레임
df_bus_card_tot = pd.DataFrame()

### 아래 데이터 통합 프로그래밍 하기...
for i in range(0, 80, 1):
    ### 1. 파일 읽어들이기
    file_path = "./01_data_원본/trfcard ({})/trfcard.csv".format(i)
    df_bus_card_org = pd.read_csv(file_path)
    
    file_path = "./01_data_원본/trfcard ({})/trfcard_columns.xlsx".format(i)
    df_bus_card_col_org = pd.read_excel(file_path,
                                        header = 2,
                                        usecols = "B:C")
    
    ### 영문 컬럼명을 한글로 변환하기
    # 딕셔너리 만들기 - 1방식
    df_bus_card_col_new_dict = {}
    for k, v in zip(df_bus_card_col_org["컬럼명 (영문)"],
                    df_bus_card_col_org["컬럼명 (한글)"]) :
        df_bus_card_col_new_dict[k] = v

    #딕셔너리 만들기 - 2방식
    df_bus_card_col_new_dict = \
                dict(zip(df_bus_card_col_org["컬럼명 (영문)"],
                         df_bus_card_col_org["컬럼명 (한글)"]))
    
    # 원본 데이터프레임의 컬럼명을 한글로 변환하기
    df_bus_card_kor = \
            df_bus_card_org.rename(columns = df_bus_card_col_new_dict)
    
    ### 체류시간(분) 컬럼 추가하기
    # 승차시각 및 하차시각 타입변환하가 : 문자형태로...
    df_bus_card_kor = df_bus_card_kor.astype({"승차시각" : "str"})
    df_bus_card_kor = df_bus_card_kor.astype({"하차시각" : "str"})
    
    # 분석에 필요한 요소(컬럼)을 별도 데이터프레임으로 추출하기
    df_bus_cart = pd.DataFrame()

    df_bus_cart["승차시각"] = \
        pd.to_datetime(df_bus_card_kor["승차시각"])
    df_bus_cart["하차시각"] = \
        pd.to_datetime(df_bus_card_kor["하차시각"])
    df_bus_cart["승객구분"] = df_bus_card_kor["승객연령"]
    df_bus_cart["환승여부"] = df_bus_card_kor["환승여부"]
    df_bus_cart["추가운임여부"] = df_bus_card_kor["추가운임여부"]
    df_bus_cart["승차정류장"] = df_bus_card_kor["승차정류장"]
    df_bus_cart["하차정류장"] = df_bus_card_kor["하차정류장"]
    
    df_bus_cart["버스내체류시간(분)"] = \
    (df_bus_cart["하차시각"] - \
         df_bus_cart["승차시각"]).astype("timedelta64[m]").astype(int)
    
    ### 기준년도, 기준월, 기준일, 기준시간, 기준시간(분) 컬럼생성
    df_bus_cart["기준년도"] = df_bus_cart["승차시각"].dt.year
    df_bus_cart["기준월"] = df_bus_cart["승차시각"].dt.month
    df_bus_cart["기준일"] = df_bus_cart["승차시각"].dt.day
    df_bus_cart["기준시간"] = df_bus_cart["승차시각"].dt.hour
    df_bus_cart["기준시간(분)"] = df_bus_cart["승차시각"].dt.minute
    
    ### 통합하기
    # - ignore_index=True : 기존의 인덱스 번호에 이어서 자동 증가시키기
    df_bus_card_tot = pd.concat([df_bus_card_tot, 
                                 df_bus_cart], ignore_index=True)

# - 종료시간
end_date = datetime.today().strftime("%Y-%m-%d %H:%M:%S")

print("전체 시작~종료 시간 = ({}) ~ ({})".format(start_date, end_date))
print("전체 데이터 행수 = ", len(df_bus_card_tot))
df_bus_card_tot
```
