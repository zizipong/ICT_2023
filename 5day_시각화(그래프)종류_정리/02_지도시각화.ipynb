```python
### 지도 시각화 라리브러리 불러들이기
# 설치 명령어 : conda install -c conda-forge folium
# 위 명령을 -> 아나콘다 프롬프트에서 실행(프롬프트 새로 열기)
# 쥬피터노트북이 실행된 프롬프트는 그대로 유지
import folium
```

```python
### json 형태 처리를 위한 라이브러리 불러들이기
import json
```

```python
### 지도맵 그리기
starbucks_map = folium.Map(
    # 최초에 보여줄 지도 위치(위경도 지정) : 중심점
    location = [37.573050, 126.979189],
    # 지도 스타일 지정 : 도시형 건물 스타일(일반적으로 사용됨)
    tiles = "openstreetmap",
    # 산림 스타일
    #tiles = "Stamen Terrain",
    # 하천/도로 스타일
    #tiles = "cartodbpositron",
    # 지도 zoom 사이즈(zoom-in / zoom-out 의 스케일값)
    zoom_start = 50
)
starbucks_map
```

```python
### 스타벅스 매장위치 표시하기
for idx in seoul_starbucks.index : 
    lat = seoul_starbucks.loc[idx, "위도"]
    lng = seoul_starbucks.loc[idx, "경도"]
    #print(lat, " / ", lng)
    
    ### 지도에 표시할 마커 모양 등 스타일 지정하기
    # - 지정한 스타일로 지도맵에 추가하기
    folium.CircleMarker(
        # 마커 위치 지정
        location = [lat, lng],
        # 마커 채우기 색상 지정
        fill_color = "green",
        # 마커 테투리 색상
        color = "red",
        # 마커 테투리 사이즈
        weight = 1,
        # 마커 크기 지정
        radius = 3
    ).add_to(starbucks_map)
    
starbucks_map    
```

```python
### 매장 타입별로 마커의 색을 구분하여 표시하기 위해서는
# - 매장 타입의 데이터 확인 : 고유한 값들만..
seoul_starbucks["매장타입"].unique()
```

```python
### 지도맵 새로 만들기(위에 맵에는 이미 마커가 추가되어 있음)
starbucks_map2 = folium.Map(
    # 최초에 보여줄 지도 위치(위경도 지정) : 중심점
    location = [37.573050, 126.979189],
    # 지도 스타일 지정 : 도시형 건물 스타일(일반적으로 사용됨)
    tiles = "openstreetmap",
    # 산림 스타일
    #tiles = "Stamen Terrain",
    # 하천/도로 스타일
    #tiles = "cartodbpositron",
    # 지도 zoom 사이즈(zoom-in / zoom-out 의 스케일값)
    zoom_start = 11
)
starbucks_map2
```

```python
### 매장 타입별로 마커 색상을 구분하여 표시하기
# 'general' (gray),  'reserve'(blue), 'generalDT'(red)

### 스타벅스 매장위치 표시하기
for idx in seoul_starbucks.index : 
    lat = seoul_starbucks.loc[idx, "위도"]
    lng = seoul_starbucks.loc[idx, "경도"]
    store_type = seoul_starbucks.loc[idx, "매장타입"]
    #print(lat, " / ", lng)
    
    fillColor = ""
    size = 1
    if store_type == "general" :
        fillColor = "gray"
        size = 3
        
    elif store_type == "reserve" :
        fillColor = "blue"   
        size = 5
        
    elif store_type == "generalDT" :
        fillColor = "red"   
        size = 5    
    
    ### 지도에 표시할 마커 모양 등 스타일 지정하기
    # - 지정한 스타일로 지도맵에 추가하기
    folium.CircleMarker(
        # 마커 위치 지정
        location = [lat, lng],    
        
        # 마커 채우기 색상 지정
        #fill_color = "green",
        fill_color = fillColor,        
        fill = True,
        
        # 마커 테투리 색상
        color = fillColor,

        # 투명도
        fill_opacity = 1,
        
        # 마커 테투리 사이즈
        weight = 1,
        # 마커 크기 지정
        radius = size
    ).add_to(starbucks_map2)
    
starbucks_map2    
```
