```python
import numpy as np
```

```python
### 25% 및 75% 시점의 데이터 확인하기
# q1 = 25%
# q3 = 75%
q1, q3 = np.percentile(tips["total_bill"], [25, 75])

print("q1 = ", q1)
print("q3 = ", q3)

iqr = q3 - q1
print("iqr = ", iqr) 

### Minimum Outliers 계산하기
lower_bound = q1 - (iqr * 1.5)
print("lower_bound = ", lower_bound)  

### Miaximum Outliers 계산하기
upper_bound = q3 + (iqr * 1.5)
print("upper_bound = ", upper_bound)  


### 이상치를 벗어나는 데이터 조회하기
tips[ (tips["total_bill"] > upper_bound) | (tips["total_bill"] < lower_bound) ]
```

```python
### 25% 및 75% 시점의 데이터 확인하기
# q1 = 25%
# q3 = 75%
q1, q3 = np.percentile(tips["tip"], [25, 75])

print("q1 = ", q1)
print("q3 = ", q3)

iqr = q3 - q1
print("iqr = ", iqr) 

### Minimum Outliers 계산하기
lower_bound = q1 - (iqr * 1.5)
print("lower_bound = ", lower_bound)  

### Miaximum Outliers 계산하기
upper_bound = q3 + (iqr * 1.5)
print("upper_bound = ", upper_bound)  


### 이상치를 벗어나는 데이터 조회하기
tips[ (tips["tip"] > upper_bound) | (tips["tip"] < lower_bound) ]
```

```python
### 25% 및 75% 시점의 데이터 확인하기
# q1 = 25%
# q3 = 75%
q1, q3 = np.percentile(tips["size"], [25, 75])

print("q1 = ", q1)
print("q3 = ", q3)

iqr = q3 - q1
print("iqr = ", iqr) 

### Minimum Outliers 계산하기
lower_bound = q1 - (iqr * 1.5)
print("lower_bound = ", lower_bound)  

### Miaximum Outliers 계산하기
upper_bound = q3 + (iqr * 1.5)
print("upper_bound = ", upper_bound)  


### 이상치를 벗어나는 데이터 조회하기
tips[ (tips["size"] > upper_bound) | (tips["size"] < lower_bound) ]
```

```python
#----- 위는 내가 한거, 아래는 강사님이랑 같이...

# 이상치 데이터 확인 --> 함수로 민들어보기

### df_org : 원본 데이터프레임 이름
### col_data :  이상치가 있을것으로 예상되는 컬럼(항목) 데이터
import numpy as np

def outliers_data(df_org, col_data) :
    
    # 25% 및 75% 시점의 데이터 확인하기
    q1, q3 = np.percentile(col_data, [25, 75])    
    
    # iqr 계산하기
    iqr = q3 - q1
    
    # 이상치 기준값 계산
    upper_bound = q3 + (iqr * 1.5)
    lower_bound = q1 - (iqr * 1.5)
    
    ### 이상치 데이터 추출하기
    df_temp = df_org[(col_data > upper_bound) | (col_data < lower_bound)]
    
    # index번호 추출
    rs_list = list(df_temp.index)
    
    ### 이상치 행의 갯수 및 이상치 데이터 리턴하기
    return len(df_temp), df_temp, rs_list
```

```python
### tips 데이터에서 total_bill 컬럼의 이상데이터 확인하기
len_row, rs_df, rs_list = outliers_data(tips, tips["total_bill"])
print(len_row, "개")
print(rs_list)
print(rs_df)
```

```python
### 히트맵 속성 설명
# 첫번째 값 : 피봇데이터
# annot : True(데이터값 보이기), False(데이터값 숨기기)
# fmt : d(정수값 표현), .2f(소수점값 표현, 소수점 2자리까지)
# cmap : 컬러맵(인터넷상에 매우 많은 맵이 존재함)
sns.heatmap(pivot_tips, 
            annot=True, fmt=".2f", 
            cmap= sns.light_palette("gray", as_cmap=True))
```
