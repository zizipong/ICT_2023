```python
### 우리나라에 입국하는 관광객이 가장 많은 국가 top5 조회
# - 전체비율이 가장 높은 상위 5개 국가 이름만 조회합니다.
# - 평균 집계하기... 
# 변수명 : temp
# 중국 > 일본 > 대만 > 미국 > 홍콩
temp = df.pivot_table(values = "전체비율(%)", 
                      index = "국적", 
                      aggfunc = "mean").sort_values(by="전체비율(%)", ascending = False).head(5)

temp
```

```python
### 피봇집계 형태를 -> 일반행렬의 데이터프레임 형태로 변환하기
# temp = temp.reset_index()
temp.reset_index(inplace = True)     # temp = temp.reset_index() 와 동일
```

```python
temp
```

```python
### top5 국가에 대한 관광객추이 그래프 그리기
# for문 사용..
# 그래프 제목에 해당 국가 이름 표시
# (예시 : [중국] 관광객 추이)
```

```python
# 강사님이랑 한거..

### 국적명 추출
contry_list = temp["국적"]

### 값을 이용하는 방식의 for문 사용
for contry in contry_list :
    ### 필터링
    condition = df["국적"] == contry    
    ### 필터링 결과 데이터 확인하기
    df_filter = df[condition] 

    ### 그래프 꾸미기
    ### 그래프 크기 조정
    plt.figure(figsize = (12, 4))

    ### 그래프 데이터 설정 및 그리기
    plt.plot(df_filter["기준년월"], df_filter["관광"])

    ### 그래프 제목 달기
    # 그래프 대제목
    plt.title("{} 국적의 관광객 추이".format(contry))
    # x축 제목
    plt.xlabel("기준년월")
    #y축 제목
    plt.ylabel("관광객수")

    ### x 축 범위를 구간으로 지정하여 보기 좋게 표시하기
    plt.xticks(["2010-01", "2011-01", "2012-01", "2013-01", "2014-01", 
                "2015-01", "2016-01", "2017-01", "2018-01", "2019-01"])

    ### 그래프 그리기
    plt.show()
```

```python
# 내가 한거...
for nm in (temp["국적"]) :
    condition = df["국적"] == nm
    
    df_filter = df[condition]
    
    ### 그래프 꾸미기
    ### 그래프 크기 조정    
    plt.figure(figsize = (12, 4))
    
    ### 그래프 데이터 설정 및 그리기    
    plt.plot(df_filter["기준년월"], df_filter["관광"])

    ### 그래프 제목 달기
    # 그래프 대제목
    plt.title("[" + nm + "]" + " 국적의 관광객 추이")
    # x축 제목
    plt.xlabel("기준년월")
    #y축 제목
    plt.ylabel("관광객수")
    
    
    ### x 축 범위를 구간으로 지정하여 보기 좋게 표시하기
    plt.xticks(["2010-01", "2011-01", "2012-01", "2013-01", "2014-01", 
            "2015-01", "2016-01", "2017-01", "2018-01", "2019-01"])

    ### 그래프 그리기
    plt.show()    
```

```python
### 5개의 선그래프를 하나의 그래프로 표현하기
### 시각화 그래프의 레이어로 쌓는 방식으로 그려집니다. 

### 국적명 추출
contry_list = temp["국적"]


### 선컬러 지정하기
# - 선의 갯수와 동일한 갯수로 리스트로 지정
color_list = ["red", "blue", "green", "orange", "gray"]

### 그래프 크기 조정
plt.figure(figsize = (12, 4))


### 값을 이용하는 방식의 for문 사용
# for contry in contry_list :
for contry, color in zip(contry_list, color_list) :
    ### 필터링
    condition = df["국적"] == contry    
    ### 필터링 결과 데이터 확인하기
    df_filter = df[condition] 

    ### 그래프 데이터 설정 및 그리기
    plt.plot(df_filter["기준년월"], df_filter["관광"], color = color, label = contry)
    
    
### 그래프 제목 달기
# 그래프 대제목
plt.title("기준년월별 국적의 관광객 추이")
# x축 제목
plt.xlabel("기준년월")
#y축 제목
plt.ylabel("관광객수")


### x 축 범위를 구간으로 지정하여 보기 좋게 표시하기
plt.xticks(["2010-01", "2011-01", "2012-01", "2013-01", "2014-01", 
            "2015-01", "2016-01", "2017-01", "2018-01", "2019-01"])

### 범례 추가하기
plt.legend()

### 그래프 그리기
plt.show()
```

```python
###  df 데이터프레임에 "년도", "월" 컬럼 생성하기
df["년도"] = df["기준년월"].str.slice(0, 4)
df["월"] = df["기준년월"].str.slice(5, 7)

# .to_list()를 붙여줘도 되고, 안붙여도 되고....
#df["년도"] = df["기준년월"].str.slice(0, 4).to_list()
#df["월"] = df["기준년월"].str.slice(5, 7).to_list()

df.info()
```

```python
### 해시맵 형태로 년도 및 월별 데이터 상세 시각화
### 처리 순서 : 피봇테이블처리 > 해시맵 시각화
```

```python
### 중국 데이터를 이용해서 해시맵 시각화 해보기
# 중국 데이터 추출하기, 변수명 : df_filter
df_filter = df[(df["국적"] == "중국")]
df_filter
```

```python
### 년도 및 월별 관광객수를 이용하여 피봇테이블 집계하기
df_pivot = df_filter.pivot_table(values = "관광",
                      index = "년도",
                      columns = "월",
                      aggfunc = "mean")
df_pivot
```

```python
### 해시맵 시각화를 위한 시각화 라이브러리 불러들이기
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
### 히트맵 그리기
# 그래프 크기 설정
plt.figure(figsize = (16, 10))

### 히트맵 그래프 그리기
sns.heatmap(df_pivot, annot = True, fmt = ".0f", cmap = "rocket_r") #annot = True, False 차이 

# 제목
plt.title("년/월별 중국 관광객 상세")
# 그래프 표현
plt.show()
```

```python
### top5 전체 히트맵 그리기

for cntry in contry_list : 
    df_filter = df[(df["국적"] == cntry)] 

    ### 년도 및 월별 관광객수를 이용하여 피봇테이블 집계하기
    df_pivot = df_filter.pivot_table(values = "관광",
                          index = "년도",
                          columns = "월",
                          aggfunc = "mean")
 
    # 그래프 크기 설정
    plt.figure(figsize = (16, 10))

    ### 히트맵 그래프 그리기
    sns.heatmap(df_pivot, annot = True, fmt = ".0f", cmap = "rocket_r") #annot = True, False 차이 

    # 제목
    plt.title("년/월별 {} 관광객 상세".format(cntry))
    # 그래프 표현
    plt.show()
```
