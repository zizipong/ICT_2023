```python
import pandas as pd 
```

```python
### 데이터 불러들이기
# - bicycle_dup.csv 파일 읽어들이기
# - 데이터프레임변수명 : df

file_path = "./data/bicycle_dup.csv"
df =  pd.read_csv(file_path, encoding='euc-kr')

df
```

```python
df.info()
```

```python
### 중복데이터 필터링 하기
# 한개 컬럼에 대한 중복데이터 필터링
# - 중복된 값 중 첫번째 중복을 제외한 모든 중복은 True
# - 만약 중복데이터를 제거하려면, True인 행만 제거...
df.duplicated(["이용거리"])
"""
홍길동   --> False
홍길동   --> True
홍길동1  --> False
홍길동   --> True
"""
```

```python
### 중복데이터 처리방법
# - 행 삭제...
# - 컬럼 삭제는 의미 없음
# - 삭제시 실제 중복이 일어나는 데이터셋의 성격인지 확인 필요
```

```python
# 이용거리 데이터에서 중복이 있는(True) 데이터 조회하기
df[df.duplicated(["이용거리"])]
```

```python
### 이용거리 및 나이를 기준으로 중복데이터 검색하기
len(df[df.duplicated(["이용거리","나이"])])
```

```python
### keep = "first"  :  첫번째 중복만 False로, 뒤쪽은 모두 True로 ~~~~~~
df.duplicated(["이용거리"], keep="first")
```

```python
### keep = "last" : 마지막 중복만 False로, 앞쪽은 모두 True로
df.duplicated(["이용거리"], keep="last")
```

### 중복데이터 삭제하기

```python
### 첫번째 중복만 남기고 나머지 중복은 삭제하기
# - 중복된 행 중에 True인 행들이 모두 삭제됨
df.drop_duplicates(["이용거리"], keep="first")
```

```python
df.drop_duplicates(["이용거리"], keep="last")
```
